```python
import pandas as pd
import numpy as np
import requests as re
from bs4 import BeautifulSoup
from time import sleep
from random import randint
```

```python
movieTitle =[]
movieRating = []
movieReleaseYear = []
movieRuntime= []
movieGenre= []
votes=[]
gross=[]
metaScore=[]
```

```python
# pages = np.arange(1,487,100)
```

```python
url ='https://www.imdb.com/list/ls046196709/?sort=list_order,asc&st_dt=&mode=detail&page='
for page in range(1, 6):
    page = re.get(url+str(page))
    soup = BeautifulSoup(page.text, 'html.parser')
    movie_data = soup.find_all('div',class_='lister-item mode-detail')
    sleep(randint(2,8))
    for data in movie_data:
        dat = data.h3.a.text
        movieTitle.append(dat)

        yearOfRelease = data.h3.find('span', class_='lister-item-year').text.replace('(','').replace(')','').replace('I','')
        movieReleaseYear.append(yearOfRelease)

        movie_rating = data.find('span', class_='ipl-rating-star__rating').text
        movieRating.append(movie_rating)

        movie_time = data.find('span', class_='runtime').text.replace('min','')
        movieRuntime.append(movie_time)

        movie_genre = data.find('span', class_='genre').text.replace('\n','').replace(' ','')
        movieGenre.append(movie_genre)   

        num_votes_gross = data.find_all('span', attrs = {'name':'nv'})
        num_votes= num_votes_gross[0].text
        votes.append(num_votes)

        num_gross = num_votes_gross[1].text if len(num_votes_gross) > 1 else '###'
        gross.append(num_gross)

        meta = data.find('span', class_='metascore').text.replace(' ','') if data.find('span', class_='metascore') else '###'
        metaScore.append(meta)
```

```python
df = pd.DataFrame({'name':movieTitle, 'rating':movieRating, 'release-year':movieReleaseYear, 'runtime':movieRuntime, 'vote':votes,'gross':gross, 'metascore':metaScore})
```
